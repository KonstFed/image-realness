```python
import matplotlib.pyplot as plt
import torch
from PIL import Image
from glob import glob
import os
```

```python
from infer import VGG
import torch

model = VGG("weights/best.pt")
```

```python
def test(path: str):
    filenames = os.listdir(path)
    fig, axs = plt.subplots(nrows=len(filenames), ncols=2, figsize=(20, 40))
    for i, file in enumerate(filenames):
        file_path = os.path.join(path, file)
        img = Image.open(file_path)
        img = img.convert("RGB")
        is_real = model.forward(img)
        axs[i][0].imshow(img)
        axs[i][1].imshow(model.cropped_img_)
        axs[i][0].set_title(is_real)
    plt.show()
        
test("./test_images")
```
